```python
covid_table = pd.read_csv("/kaggle/input/corona-virus-report/covid_19_clean_complete.csv")
a = covid_table.columns                       #Nome das colunas
print(len(covid_table.columns))           #quantidade de colunas 

#tipo de cada dado da coluna
i = 0
for i in range(len(covid_table.columns)):
    print("A coluna {} é um dado tipo {}.". format(a[i] ,type(covid_table[a[i]][1])))
    
```

```python
#Index(['Province/State', 'Country/Region', 'Lat', 'Long', 'Date', 'Confirmed',
#       'Deaths', 'Recovered', 'Active', 'WHO Region'],

covid_table1 = covid_table.fillna(0)                                                 # Substituir todos os valores NaN por 0
covid_table0 = covid_table1[covid_table['Province/State'].notna()]                   # Tira os valores vazios 
covid_table0 = covid_table0.sort_values(by=['Confirmed'],ascending = False)          # Ordena os valores de Maior para Menor
covid_table0 = covid_table0.drop_duplicates('Province/State')                        #Tira as provincias repetidas
#covid_table0 

#covid_table0 = covid_table0.sort_values(by=['Confirmed'],ascending = False)
d = ['Province/State', 'Confirmed', 'Active', 'Deaths', 'Recovered']
top5 = covid_table0[d]
print("\033[1mTOP 5 PROVINCIAS COM MAIS CASOS DE COVID\033[0m\n")
top5.head(5)
```

## Função para juntar a coluna provincia (onde notna) com a coluna pais 

```python
def juntando(linha):                                     #Def uma fução com uma linha de argumento que será utilizado pelo aplly
    if pd.notna(linha['Province/State']):                #Para cada linha da coluna Province/State será verdadeiro caso NÂO for NaN
        linha['Country/Region'] = linha['Country/Region'] + '_' + linha['Province/State'] # Junta as duas colunas
    else:
        linha['Country/Region'] = linha['Country/Region'] #Caso for FALSO não vai juntar as duas colunas 
    
    return linha                                           
 

covid_new = covid_table.apply(juntando, axis=1)           #Utilizando o apply para fazer o nome dataframe
covid_new
```

```python
m = ['Deaths', 'Country/Region']
covid_compere = covid_table.sort_values(by=['Deaths'],ascending = False)
covid_compere = covid_compere[m]  #Dataframe com o numero de mortos por pais ou região
```

```python
#xu = covid_compere[covid_compere['Country/Region'] == 'Yemen']
#xu = xu.sort_values(by=['Deaths'],ascending = False)
covid_compere0 = covid_compere.drop_duplicates('Country/Region')
```

```python
covid_compere0
```

```python
World_data = pd.read_csv("/kaggle/input/corona-virus-report/worldometer_data.csv")
World_data
#Tome apenas as informações de população, país e continentes
v = ['Population', 'Country/Region', 'Continent']
World_data = World_data[v]
World_data0 = World_data.drop_duplicates('Country/Region')
World_data0

#Juntando os dois dataframes!!!
#covid_compere.drop_duplicates('Country/Region')

table_together = World_data0.merge(covid_compere0, on='Country/Region')
table_together
```

### Agrupando a tabela final por continentes e mostrando qual tem mais mortes

```python
#table_together.drop_duplicates('Continent')
#[South America, Asia, Europe, Africa, North America, Australia/Oceania]

table_continent = table_together.groupby('Continent').sum()
colunas_para_excluir = ['Country/Region']
table_continent = table_continent.drop(colunas_para_excluir, axis=1)
table_continent
topp5 = table_continent.sort_values(by=['Deaths'],ascending = False)

print("TOP 5 CONTINENTES COM MAIS MORTES")
topp5.head(5)
```

### Morte por Milhão

```python
#Crie uma nova coluna contendo o número de mortes por milhão dentro dos
#continentes (Utilize a fórmula: (mortes/população)*10^6).

topp5['Morte/Milhão'] = (topp5['Deaths']/topp5['Population'])*1000000

print("MORTE POR MILHÃO DE PESSOAS EM CADA CONTINENTE")
topp5
```
